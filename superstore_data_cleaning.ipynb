```python

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

df.dtypes
```

```python
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days


df['Delivery Days'].head(10)
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month


df[['Order Date', 'Order Year', 'Order Month']].head(10)
```

```python

print("====== SUPERSTORE SALES ANALYSIS ======")
print()
print("1. Total Sales: $", round(df['Sales'].sum(), 2))
print()

# Sales by Region
print("2. Sales by Region:")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False))
print()

# Sales by Category
print("3. Sales by Category:")
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False))
print()

# Sales by Segment
print("4. Sales by Segment:")
print(df.groupby('Segment')['Sales'].sum().sort_values(ascending=False))
print()

# Top 5 Products
print("5. Top 5 Products by Sales:")
print(df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head())
print()

# Sales by Year
print("6. Sales by Year:")
print(df.groupby('Order Year')['Sales'].sum().sort_values(ascending=False))
print()

# Average Delivery Days
print("7. Average Delivery Days:", round(df['Delivery Days'].mean(), 1))
print()

# Average Delivery by Ship Mode
print("8. Average Delivery Days by Ship Mode:")
print(df.groupby('Ship Mode')['Delivery Days'].mean().sort_values())
print()
```
